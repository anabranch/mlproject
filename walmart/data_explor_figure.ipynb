```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import seaborn as sns
```

```python
raw = pd.read_csv("data/train.csv", dtype = {'FinelineNumber': str, 'Weekday': str, 'UPC': str})
```

```python
plt.figure()
```

```python
trip_frequency = raw.groupby('TripType').size()
trip_frequency
```

```python
trip_frequency.plot(kind = 'bar', color = 'r')
plt.title("TripType distribution in training set")
plt.show()
```

```python
visit_number_groupby = raw.groupby('VisitNumber')
triptype = visit_number_groupby.mean()[['TripType']]
scan_count_per_visit = visit_number_groupby.sum()[['ScanCount']]
scan_count_per_visit['TripType'] = triptype.astype(int)
```

```python
gb_scan_count = scan_count_per_visit.groupby('TripType')
gb_scan_count.median().plot(kind = 'bar')
plt.show()
```

```python
gb_scan_count.mean().plot(kind = 'bar')
plt.show()
```
